```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from IPython.display import clear_output
import time

def track_gray_box(video_filename):
    # Read the input video
    video_cap = cv2.VideoCapture(video_filename)

    # Get video properties
    fps = video_cap.get(cv2.CAP_PROP_FPS)
    frame_size = (int(video_cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(video_cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

    # Create a VideoWriter object for the output video
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    output_filename = 'tracked_gray_box_video.avi'
    video_writer = cv2.VideoWriter(output_filename, fourcc, fps, frame_size)

    while video_cap.isOpened():
        ret, frame = video_cap.read()
        if not ret:
            break

        # Convert the frame to grayscale
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Find contours in the frame
        contours, _ = cv2.findContours(gray_frame, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        for contour in contours:
            # Check if the contour area is within a certain range (adjust as needed)
            area = cv2.contourArea(contour)
            if 500 < area < 5000:
                # Get the bounding box around the contour
                x, y, w, h = cv2.boundingRect(contour)

                # Draw a red bounding box around the gray box
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)

                # Add label "gray box"
                cv2.putText(frame, "Gray Box", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

        # Write the frame to the output video
        video_writer.write(frame)

        # Display the frame
        cv2_imshow(frame)

        # Clear the output to avoid multiple windows in Colab
        clear_output(wait=True)

        # Pause for a short duration to control the playback speed
        time.sleep(0.1)

    # Release resources
    video_cap.release()
    video_writer.release()

# Example usage: replace 'input_video.avi' with the path to your input video
track_gray_box(video_filename)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from IPython.display import clear_output
import time

def track_gray_box(video_filename):
    # Read the input video
    video_cap = cv2.VideoCapture(video_filename)

    # Get video properties
    fps = video_cap.get(cv2.CAP_PROP_FPS)
    frame_size = (int(video_cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(video_cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

    # Create a VideoWriter object for the output video
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    output_filename = 'tracked_gray_box_video_with_coordinates.avi'
    video_writer = cv2.VideoWriter(output_filename, fourcc, fps, frame_size)

    while video_cap.isOpened():
        ret, frame = video_cap.read()
        if not ret:
            break

        # Convert the frame to grayscale
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Find contours in the frame
        contours, _ = cv2.findContours(gray_frame, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        for contour in contours:
            # Check if the contour area is within a certain range (adjust as needed)
            area = cv2.contourArea(contour)
            if 500 < area < 5000:
                # Get the bounding box around the contour
                x, y, w, h = cv2.boundingRect(contour)

                # Calculate the center coordinates
                center_x = x + w // 2
                center_y = y + h // 2

                # Draw a red bounding box around the gray box
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)

                # Add label "gray box" and coordinates
                label_text = f"Gray Box: ({center_x}, {center_y})"
                cv2.putText(frame, label_text, (x, y - 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

        # Write the frame to the output video
        video_writer.write(frame)

        # Display the frame
        cv2_imshow(frame)

        # Clear the output to avoid multiple windows in Colab
        clear_output(wait=True)

        # Pause for a short duration to control the playback speed
        time.sleep(0.1)

    # Release resources
    video_cap.release()
    video_writer.release()

# Example usage: replace 'input_video.avi' with the path to your input video
track_gray_box(video_filename)
```
